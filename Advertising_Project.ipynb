```python
df1 = df.drop(['Ad Topic Line','City','Country','Timestamp'], axis=1)
```

```python
df1.head()
```

```python
df1['Clicked on Ad'].value_counts()
```

```python
X = df1[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y = df1['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ran = RandomForestClassifier(n_estimators=200)
```

```python
ran.fit(X_train, y_train)
```

```python
ran.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(RandomForestClassifier(), {
    'n_estimators': [50,100,150,200]
}, cv=5, return_train_score=False)
clf.fit(X_train, y_train)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_n_estimators','rank_test_score','mean_test_score']]
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'), {
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5, return_train_score=False)
clf.fit(X_train, y_train)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
svc = SVC(gamma='auto', kernel='linear', C=10)
```

```python
svc.fit(X_test, y_test)
```

```python
svc.score(X_test,y_test)
```

```python
pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(pred, y_test)
cm
```

```python
cf_report = classification_report(pred, y_test)
print(cf_report)
```
